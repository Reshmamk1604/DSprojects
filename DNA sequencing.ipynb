```python
import pandas as pd
```

```python
data = pd.read_csv('human_data.txt',sep = '\t')
```

```python
data.columns
```

```python
data['class'].value_counts()
```

```python
def slicing(sequence,size=6):  #k-mers concept
    seq=[]
    print('seq:',sequence)
    for i in range(len(sequence)-6 +1):
        seq.append(sequence[i:i+6])        
    return seq    
```

```python
data['kmer']= data.apply(lambda x:slicing(x['sequence']),axis=1)
```

```python
data.drop('sequence',axis = 1,inplace=True)
```

```python
type(data['kmer'])
```

```python
kmer_list =[]
for i in range(len(data['kmer'])):
    plain = data['kmer'][i]
    kmer_list.append(plain)
    plain = ''
    
```

```python
kmer_list = list(data['kmer'])
for i in range(len(kmer_list)):
    kmer_list[i] = ' '.join(kmer_list[i])
```

```python
y_data = data.iloc[:, 0].values
```

```python
y_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vect = CountVectorizer(ngram_range=(4,4))
x = vect.fit_transform(kmer_list)
x.shape
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, 
                                                    y_data, 
                                                    test_size = 0.20, 
                                                    random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
```

```python
X_train = X_train.todense()
X_test = X_test.todense()
# todense is used because of this error
#TypeError: A sparse matrix was passed, but dense data is required.
#Use X.toarray() to convert to a dense numpy array.
```

```python
model.fit(X_train, y_train)
# make predictions
```

```python
predicted = model.predict(X_test)
predicted
```

```python
classifier = MultinomialNB(alpha=0.1)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
print("Gaussian NB metrics\n")
print("Confusion matrix\n")
print(pd.crosstab(pd.Series(y_test, name='Actual'), pd.Series(y_pred, name='Predicted')))
def get_metrics(y_test, predicted):
    accuracy = accuracy_score(y_test, predicted)
    precision = precision_score(y_test, predicted, average='weighted')
    recall = recall_score(y_test, predicted, average='weighted')
    f1 = f1_score(y_test, predicted, average='weighted')
    return accuracy, precision, recall, f1
accuracy, precision, recall, f1 = get_metrics(y_test, y_pred)
print("accuracy = %.3f \nprecision = %.3f \nrecall = %.3f \nf1 = %.3f" % (accuracy, precision, recall, f1))
```
